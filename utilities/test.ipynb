## Improving zscore

```python
df = pd.DataFrame({'gene': ['A1BG', 'PRB1', 'FOXA1', 'TULP2'] * int(10000/4), 
                   'id': np.random.randint(100, size=10000),
                   'col3': np.random.randint(100, size=10000), 
                   'col4': np.random.randint(100, size=10000)})
df.set_index('gene', inplace=True)
df2 = df.copy()
```

```python
%timeit uf.zscore(df, axis='row')
```

```python
%timeit uf2.zscore(df2, axis='row',epsilon=0)
```

```python
df
```

```python
df2
```

```python
assert(df.equals(df2))
```
